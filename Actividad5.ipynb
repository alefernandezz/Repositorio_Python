```python
n1=int(input("Ingrese un numero por favor "))
n2=int(input("Ingrese un segundo numero "))
print("*****MENU*****")
#opciones del menu
print("1-Realizar suma de los numeros ingresados")
print("2-Realizar la resta de los numeros ingresados")
print("3-Realizar la multiplicacion de los numeros ingresados")
print("4-Interrumpir programa")
opt=int(input("Ingrese la opcion deseada "))
while (opt == 1 or opt == 2 or opt == 3 or opt == 4):
  if (opt == 1):
    suma=n1+n2
    print("La suma de los numeros ingresados es ",suma)
    break
  if (opt == 2):
      resta=n1-n2
      print("La resta de los numeros ingresados es ",resta)
      break
  if (opt == 3):
      mult=n1*n2
      print("La multiplicacion de los numeros ingresados es ",mult)
      break
  if (opt == 4):
      print("Adios")
      break
else:
  print("La opcion ingresada no es valida")
```

```python
#Ejercicio 4
cantNum = int(input("Ingrese cantidad de numeros que quiere introducir: "))
n = cantNum
#Cree variable n porque cantNum al final va a ser igual a 0 por lo  tanto tengo una division por cero
total = 0
while cantNum != 0:
    a = int(input("Ingrese numero: "))
    total = a + total
    cantNum -= 1
media = total/n
print("La media de los numeros ingresados es:",media)
```
