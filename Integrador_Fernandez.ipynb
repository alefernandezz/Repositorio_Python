```python
#Ejercicio 2 - POO
# Crear la clase Bicicleteria.
# Declarar los atributos bicicletas, ganancias y cantidad_de_ventas.
# Implementar el método vender_bicicleta() de la clase Bicicleteria.
# Implementar el método comprar_bicicleta().
# Crear la clase Bicicleta.
# Declarar los atributos de la clase Bicicleta: nro_de_serie, modelo, año y precio.
# Implementar el método get_precio().
# Implementar el método set_precio().
# Implementar el método get_nro_de_serie().
# Se deberán crear al menos cuatro bicicletas y una bicicleteria. Deben usarse todos los
#métodos al menos una vez
class Bicicleteria:
  def __init__(self,bicicletas=[],ganancias=0,cantidad_de_ventas=0):
    self.bicicletas=bicicletas
    self.ganancias=ganancias
    self.cantidad_de_ventas=cantidad_de_ventas
  def vender_bicicleta(self,ganancias=float,cantidad_de_ventas=int):
    self.ganancias+=ganancias
    self.cantidad_de_ventas+=cantidad_de_ventas
    print(f"Cantidad de bicis vendidas: {cantidad_de_ventas}")
    print(f"Ganancias totales obtenidas de la venta: {ganancias}")
  def comprar_bicicleta(self,bicicletas):
    self.bicicletas.append(bicicletas)
    return (f"Ahora se tienen {self.bicicletas} bicicletas en stock")

  class Bicicleta:
    def __init__(self,nro_serie,modelo,anio,precio):
      self.nro_serie=nro_serie
      self.modelo=modelo
      self.anio=anio
      self.precio=precio
    def set_precio(self,nuevo_precio):
      #Permite modificar el precio de una bici
      self.nuevo_precio=nuevo_precio
      return (f"El nuevo precio es de: ${self.nuevo_precio}")
    def get_precio(self):
      return (f"El precio de la bici es de: ${self.precio}")
    def get_nro_serie(self):
      return (f"El número de serie es: {self.nro_serie}")

bicicleteria_1=Bicicleteria(["Venzo","Vairo"],52300,2)
print(bicicleteria_1)
print(bicicleteria_1.vender_bicicleta(15000,1))
print(bicicleteria_1.comprar_bicicleta("Trek"))
#Crear 4 bicicletas
bici1=Bicicleteria.Bicicleta("AF0023","Mountain Bike",2019,45000)
bici2=Bicicleteria.Bicicleta("AF0230","Competición",2020,86700)
bici3=Bicicleteria.Bicicleta("AF0145","BMX",2021,37800)
bici4=Bicicleteria.Bicicleta("AF0301","Gravel",2020,134000)
print(bici1.get_nro_serie())
print(bici2.set_precio(75200))
print(bici3.get_precio())
print(bici4.get_nro_serie())
```

```python
#Ejercicio 4 - Analisis de Datos
import pandas as pd
import matplotlib.pyplot as plt
# Primero genero un DataFrame con los datos del fichero a utilizar
titanic = pd.read_csv('https://raw.githubusercontent.com/NotAyushXD/Titanic-dataset/master/train.csv', index_col=0)
#1 Mostrar la información de los primeros 5 pasajeros.
print('Primeros 5 pasajeros:\n', titanic.head(5))
#2 Mostrar la información de los últimos 5 pasajeros.
print('Últimos 5 pasajeros:\n', titanic.tail(5))
#3 Mostrar la cantidad total de filas y columnas.
print('Filas y columnas:', titanic.shape)
#4 ¿Cuánto dinero se recaudó en ese viaje?
Total= titanic['Fare'].sum()
print(f'La cantidad de dinero recaudada fue de: $ {Total}')
#5 ¿Cuál es la Cantidad de personas que viajaban en primera clase?
Total_first_class= titanic['Pclass'].sum()
print(f'La cantidad de pasajeros que viajaban en primera clase es de: {Total_first_class}')
#6 Graficar la cantidad de personas que viajaban por cada clase del barco.
titanic['Pclass'].value_counts().plot(kind="bar")
```
