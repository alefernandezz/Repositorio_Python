```python
numero_1=(int(input("Ingrese un numero entero ")))
numero_2=(int(input("Ingrese un numero entero")))
resto=(numero_1%numero_2)
if (resto == 0):
  print("El resultado de la division de los numeros es exacta")
else:
  print("El resultado de la division de los numeros no es exacta")
```
