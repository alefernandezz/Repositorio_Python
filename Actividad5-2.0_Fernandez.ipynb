```python
#Crear una funcion que dada una base y una altura de un rectangulo calcule su area
def area_rectangulo(b,h):
    area=b*h
    return (area)
area_rect=area_rectangulo(15,10)
print("El area del rectangulo es de: ",area_rect)
```

```python
import math
def area_circulo(rad):
    area=(rad**2)*math.pi
    return(area)
circle_area= area_circulo(5)
print("El area del circulo es de: ",circle_area)
```

```python
#Realizar una funcion llamada relacion() que a partir de dos numeros cumpla lo siguiente:
#si el primer numero es mayor que el segundo debe devolver 1
#si el primer numero es menor que el segundo debe devolver -1
#si ambos numeros son iguales debe devolver un 0
#comprueba entre 5 y 10, 10 y 5, 5 y 5
def relacion(a,b):
    if (a>b):
        result=1
    if (a<b):
        result=-1
    if (a==b):
        result=0
    return (result)
resultado=relacion(5,10)
print(resultado)
resultado=relacion(10,5)
print(resultado)
resultado=relacion(5,5)
print(resultado)
```

```python
def intermedio(a,b):
    suma=(a+b)/2
    return (suma)
interm=intermedio(-12,24)
print("El numero intermedio entre -12 y 24 es: ",interm)
```

```python
def recortar(crop,lim_inf,lim_sup):
    if (crop < lim_inf):
        nro=lim_inf
    if (crop > lim_sup):
        nro=lim_sup
    if (crop <= lim_sup and crop>= lim_inf):
        nro=crop
    return (nro)
recorte=recortar(15,0,10)
print(recorte)
```

```python
def separar(lista):
  lista_par=[]
  lista_imp=[]
  for num in lista:
    if (num%2 ==0):
      lista_par.append(num)
    else:
      lista_imp.append(num)
  return (lista_par,lista_imp)
separados=separar([2,21,17,52,78,43])
print(separados)
```
