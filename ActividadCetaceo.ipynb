```python
class Mamifero:
  def __init__(self,cantMamas,esperanzaDeVida):
    self.cantMamas=1
    self.esperanzaDeVida=esperanzaDeVida
  def describir(self,esperanzaDeVida):
    print(f"Soy un mamífero, tengo 1 mamá y {esperanzaDeVida} de esperanza de vida")
  def mamar(self):
    print("Está mamando, no molestar")
  def nacer(self):
    print("Esta naciendo!")
class AnimalMarino:
  def __init__(self,tieneBranqueas,especie):
    self.tieneBranqueas=tieneBranqueas
    self.especie=especie
  def nadar(self):
    print("Estoy nadando")
  def describirme(self,tieneBranqueas,especie):
    print(f"Soy ademas un Animal Marino {tieneBranqueas} branqueas y soy una {especie}")
class Cetaceo(Mamifero,AnimalMarino):
  def __init__(self,notas,viveEn,peso):
    self.notas=notas
    self.viveEn=viveEn
    self.peso=peso
  def showout(self,notas,viveEn,peso):
    print(f"Actualmente en {notas} residencia en {viveEn} con un peso de {peso}")
cetaceo1=Cetaceo("observacion","Atlántico Sur","30 toneladas")
cetaceo1.nacer()
cetaceo1.mamar()
cetaceo1.nadar()
cetaceo1.describir("70 años")
cetaceo1.describirme("tengo","Ballena Azul")
cetaceo1.showout("observacion","Atlántico Sur","130 toneladas")



```

```python
class Bicicleteria:
  def __init__(self,bicicletas=[],ganancias=0,cantidad_de_ventas=0):
    self.bicicletas=bicicletas
    self.ganancias=ganancias
    self.cantidad_de_ventas=cantidad_de_ventas
  def vender_bicicleta(self,ganancias=float,cantidad_de_ventas=int):
    self.ganancias+=ganancias
    self.cantidad_de_ventas+=cantidad_de_ventas
    print(f"Cantidad de bicis vendidas: {cantidad_de_ventas}")
    print(f"Ganancias totales obtenidas de la venta: {ganancias}")
  def comprar_bicicleta(self,bicicletas):
    self.bicicletas.append(bicicletas)
    return (f"Ahora se tienen {self.bicicletas} bicicletas en stock")

  class Bicicleta:
    def __init__(self,nro_serie,modelo,anio,precio):
      self.nro_serie=nro_serie
      self.modelo=modelo
      self.anio=anio
      self.precio=precio
    def set_precio(self,nuevo_precio):
      #Permite modificar el precio de una bici
      self.nuevo_precio=nuevo_precio
      return (f"El nuevo precio es de: ${self.nuevo_precio}")
    def get_precio(self):
      return (f"El precio de la bici es de: ${self.precio}")
    def get_nro_serie(self):
      return (f"El número de serie es: {self.nro_serie}")

bicicleteria_1=Bicicleteria(["Venzo","Vairo"],52300,2)
#print(bicicleteria_1)
print(bicicleteria_1.vender_bicicleta(15000,1))
print(bicicleteria_1.comprar_bicicleta("Trek"))
#Crear 4 bicicletas
bici1=Bicicleteria.Bicicleta("AF0023","Mountain Bike",2019,45000)
bici2=Bicicleteria.Bicicleta("AF0230","Competición",2020,86700)
bici3=Bicicleteria.Bicicleta("AF0145","BMX",2021,37800)
bici4=Bicicleteria.Bicicleta("AF0301","Gravel",2020,134000)
print(bici1.get_nro_serie())
print(bici2.set_precio(75200))
print(bici3.get_precio())
print(bici4.get_nro_serie())
```
